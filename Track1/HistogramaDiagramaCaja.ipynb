```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/MyDrive/PTCienciaDatos/PandasParaManipulacionDeDatos/catsvdogs.xlsx'
df = pd.read_excel(filename)
df.head()
```

1. Trazado de histograma del porcentaje de dueños de perros.

```python
plt.style.use('classic')
df['Percentage of Dog Owners'].hist(fc='#708CFF', edgecolor='k', bins=7)
plt.xlabel('Porcentaje de dueños de perros', fontsize=16, fontweight='bold')
plt.ylim(top=15);
```

2. Trazado de histograma del porcentaje de dueños de gatos.

```python
plt.style.use('classic')
df['Percentage of Cat Owners'].hist(fc='#FEA617', edgecolor='k', bins=15)
plt.xlabel('Porcentaje de dueños de gatos', fontsize=16, fontweight='bold')
plt.ylim(top=15);
```

3. Trazado del histograma del porcentaje de due;os de perros y gatos.

```python
plt.style.use('classic')
df['Percentage of Dog Owners'].hist(fc='#708CFF', edgecolor='k', bins=7, label='Perros')
df['Percentage of Cat Owners'].hist(fc='#FEA617', edgecolor='k', bins=15, label = 'Gatos')
plt.xlabel('Porcentaje de dueños de animales', fontsize=16, fontweight='bold')
plt.ylabel('Conteo', fontsize=16, fontweight='bold')
plt.ylim(top=15)
plt.legend();
```

De este gráfico podemos decir a leve rasgo, que las personan con mayor frecuencia prefieren tener perros antes que gatos.

4. Trazado de diagramas de caja en base al promedio de perros y gatos por casa

```python
plt.style.use('seaborn')
boxplot = plt.boxplot([df['Mean Number of Cats'], df['Mean Number of Dogs per household']], 
            notch=True,
            widths=.7,
            patch_artist=True,
            medianprops=dict(linestyle='-', linewidth=2, color='yellow'),
            labels=['Gatos','Perros'],
            boxprops=dict(linestyle='-', linewidth=1, color='k', facecolor='#FCD05A', alpha=.4))
boxplot1 = boxplot['boxes'][1]
boxplot1.set_facecolor('#0E33A7')
plt.ylim(top=2.8)
plt.ylabel('Promedio de animales por casa', fontsize=16)
```

5. Del gráfico obtenido podemos mencionar que de aquellos hogares que gustan de tener gatos, en promedio tienen 2 gatos, mientros que aquellos hogares que gustan de los perros, en promedio tienen menos de 2 perros.
