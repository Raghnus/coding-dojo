```python
# imports
import pandas as pd
from sqlalchemy import create_engine
```

```python
# montar datos
from google.colab import drive
drive.mount('/content/drive')
```

```python
# cargar datos: ¡cambia la ruta para que coincida con tu unidad!
school = pd.read_csv('/content/drive/MyDrive/PTCienciaDatos/SQL/School_Details.csv')
# Asegúrate de que estén cargados correctamente
school.head(2)
```

```python
# crear motor sqlite
engine = create_engine('sqlite:///school.db', echo=True)
# crear conexión con el motor
conn = engine.connect()
```

```python
# agregar marco de datos como una tabla en sqlite
sqlite_table = "school_details"
school.to_sql(sqlite_table, conn, if_exists='fail')
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, SchoolLevel, Website
FROM school_details
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, "DistrictHQDistance (in Km)"
FROM school_details
WHERE "DistrictHQDistance (in Km)" > 100
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, SchoolLevel,"DistrictHQDistance (in Km)"
FROM school_details
WHERE SchoolLevel like "JHS"
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, SchoolLevel,"DistrictHQDistance (in Km)"
FROM school_details
WHERE EstablishmentName like "%JUNIOR%HIGH%"
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, SchoolLevel,"DistrictHQDistance (in Km)"
FROM school_details
WHERE EstablishmentName like "C%"
ORDER BY EstablishmentName
LIMIT 10
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName
FROM school_details
WHERE LocatedInRuralOrUrban like "Urban Area"
ORDER BY EstablishmentName desc
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName AS Nombre
FROM school_details
WHERE LocatedInRuralOrUrban like "Urban Area"
ORDER BY EstablishmentName desc
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```
