```python
# cargar datos: ¡cambia la ruta para que coincida con tu unidad!
school = pd.read_csv('/content/drive/MyDrive/PTCienciaDatos/SQL/School_Details.csv')
# Asegúrate de que estén cargados correctamente
school.head(2)
```

```python
# cargar datos: ¡cambia la ruta para que coincida con tu unidad!
facility = pd.read_csv('/content/drive/MyDrive/PTCienciaDatos/SQL/School_Facility_Details.csv')
# Asegúrate de que estén cargados correctamente
facility.head(2)
```

```python
# crear motor sqlite
engine = create_engine('sqlite:///school.db', echo=True)
# crear conexión con el motor
conn = engine.connect()
```

```python
# agregar marco de datos como una tabla en sqlite
sqlite_table = "school_details"
school.to_sql(sqlite_table, conn, if_exists='fail')
```

```python
# agregar marco de datos como una tabla en sqlite
sqlite_table_2= "school_facility"
facility.to_sql(sqlite_table_2, conn, if_exists='fail')
```

```python
sql = """
select avg("BACDistance (in Km)") as "avg_bac_distancia"
from school_details
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
select avg("BACDistance (in Km)") as "avg_bac_distancia"
from school_details
group by SchoolLevel
order by "avg_bac_distancia" desc
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
select avg("BACDistance (in Km)") as "avg_bac_distancia"
from school_details
group by SchoolLevel
having "avg_bac_distancia" < 10
order by "avg_bac_distancia" desc

"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
select *
from school_details s INNER JOIN school_facility f
limit 2

"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
select s.EstablishmentName
from school_details s INNER JOIN school_facility f
where s.SchoolLevel LIKE "PS" and f.IsLibraryAvailable like "Available"
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df
```

```python
sql = """
select s.EstablishmentName
from school_details s INNER JOIN school_facility f
where s.SchoolLevel LIKE "PS" and f.IsLibraryAvailable like "Available"
order by s.EstablishmentName desc
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df
```
