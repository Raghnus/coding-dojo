```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
from google.colab import drive
```

```python
df_train=pd.read_csv('/content/drive/MyDrive/PTCienciaDatos/DeepLearning/RNC/fashion-mnist_train.csv')
df_train.head()
```

```python
df_test=pd.read_csv('/content/drive/MyDrive/PTCienciaDatos/DeepLearning/RNC/fashion-mnist_test.csv')
df_test.head()
```

```python
X_train=df_train.drop(columns={'label'})
y_train=df_train['label']
X_test=df_test.drop(columns={'label'})
y_test=df_test['label']
```

```python
def pixeles(X_train, y_train, X_test, y_test):
 X_train = X_train/255
 X_test = X_test/255
 X_train = np.array(X_train)
 X_test = np.array(X_test)
 X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
 X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
 y_train = to_categorical(y_train, num_classes = 10)
 y_test = to_categorical(y_test, num_classes = 10)
 return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = pixeles(X_train,y_train, X_test, y_test)
```

```python
forma=X_train.shape[1:4]
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters = 10, 
                 kernel_size = 5,
                 input_shape = forma)) 
# Pooling layer
model.add(MaxPooling2D(pool_size = 2)) # Tamaño de pooling
# Capa de aplanamiento
model.add(Flatten())
# Capa de salida
model.add(Dense(10, # Cuantas posibilidades de salida tenemos
                activation = 'softmax')) # ¿Qué función de activación estás usando?
```

```python
# Paso 2: compilar
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['acc'])
```

```python
# Paso 3: ajustar nuestro modelo
history = model.fit(X_train, y_train,
                    validation_data = (X_test, y_test), 
                    epochs=10)
```
